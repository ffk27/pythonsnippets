```python
import math
import matplotlib.pyplot as plt

#      c
#     /|
#    / |
#   /  |
#  /___|
# a     b
#

R = 6371e3 * 1.0 # straal aarde

ab = 6536
deg = 180*ab/(math.pi*R)
h1 = 4
h2 = 0.25
bc = 9
x=(h2-h1)/((bc-h1)/ab)
#print(round(-x+ab))

def dishorizon(h):
    return round(math.sqrt((R+h)**2 - R**2))

def kromming(x):
    return -(math.sqrt(R**2 + x**2) - R)

def g(x):
    return -((bc-(h1+kromming(ab))) / ab) * x + bc

afstanden = [i for i in range(0, 20000, 100)]
ys = [kromming(x) for x in afstanden]
plt.plot(afstanden, [kromming(x) for x in afstanden])
plt.plot(afstanden, [g(x) for x in afstanden])
bc = 11
plt.plot(afstanden, [g(x) for x in afstanden])
bc = 15
plt.plot(afstanden, [g(x) for x in afstanden])
plt.plot([ab], [4+kromming(ab)], marker='o', markersize=3, color="red")
plt.xlabel('Afstand (m)')
plt.ylabel('Hoogte (m)')
plt.xlim(0,afstanden[len(afstanden)-1])
plt.show()
```
