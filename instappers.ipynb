```python
import csv
import matplotlib.pyplot as plt
import os

data = []
with open(os.path.join(os.getcwd(), 'data/instappers.csv')) as csvfile:
    read = csv.reader(csvfile)
    for row in read:
        for i in range(0,len(row)):
            try:
                row[i] = int(row[i])
            except:
                pass
        data.append(row)

def grafiek(stations):
    for s in stations:
        plt.plot(range(2013,2019),s[1:len(s)],label=s[0])
    totaal = [sum([s[i] for s in stations]) for i in range(1, 7)]
    #groei1318p = [s[len(s)-1] / s[1] * 100 - 100 for s in stations]
    plt.plot(range(2013,2019),totaal, label='Totaal')
    plt.legend(loc="upper left")
    plt.show()

def getStations(naam=None, namen=None):
    if (namen):
        return [[s for s in data if s[0] == n][0] for n in namen]
    return [s for s in data if naam in s[0]]
#grafiek("Amsterdam")

#almere = getStations('Almere')
sLijnA = getStations(namen=['Leeuwarden', 'Grou-Jirnsum', 'Akkrum', 'Heerenveen', 'Wolvega', 'Steenwijk', 'Meppel'])
sLijnB = getStations(namen=['Groningen', 'Groningen Europapark', 'Haren', 'Assen', 'Beilen', 'Hoogeveen', 'Meppel'])
sLijnE = getStations(namen=['Venlo', 'Blerick', 'Horst-Sevenum', 'Deurne', 'Helmond Brouwhuis', 'Helmond', 'Helmond \'t Hout', 'Helmond Brandevoort'])
grafiek(sLijnA)
```
