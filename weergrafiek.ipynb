```python
import csv, os
import matplotlib.pyplot as plt
from lib.codes import *
from lib.lib import maanden

def loadCSVFile(filename):
    with open(os.path.join(os.getcwd(), filename)) as csvfile:
        return [[v.strip() for v in row] for row in csv.reader(csvfile, delimiter=',')]

def loadWeatherData(filename):
    data = []
    unformated = loadCSVFile(filename)
    for i in range(1, len(unformated)):
        row = unformated[i]
        rowdata = []
        for i in range(0, len(row)):
            celldata = None
            if (row[i]):
                celldata = datatypes[i](row[i])
            rowdata.append(celldata)
        data.append(rowdata)
    return data

weahterData = loadWeatherData('data/his_42.csv')

def plotMonth(year, month):
    plt.plot([data[4] for data in weahterData if data[1][0]==year and data[1][1]==month and data[4]])
    plt.title('Temperatuurverloop ' + maanden[month-1] + ' ' + str(year))
    plt.show()

plotMonth(1845,3)
# years = list({data[1][0] for data in weahterData})
# months = list({data[1][1] for data in weahterData})

# for year in years:
#     avgmonth=[]
#     for month in months[0:3]:
#         temps = [data[4] for data in weahterData if data[1][0]==year and data[1][1]==month and data[4]]
#         if (len(temps) > 0):
#             meantemp = round(sum(temps)/len(temps), 1)
#             if meantemp < 0: print(year, month, meantemp)
#             avgmonth.append(sum(temps) / len(temps))
    #plt.plot(avgmonth, label=str(year) + '-' + str(month))
    #yeartemps = [data[4] for data in weahterData if data[1][0]==years[i]]
    #plt.plot(yeartemps, label=years[i])
#plt.show()
```
